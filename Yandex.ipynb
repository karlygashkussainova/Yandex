```python
def candies(lst):
    temp=set(lst)
    result={}
    for i in temp:
        result[i]=lst.count(i)
    #print (result)
    new_lst = []

    for i in result.keys():
        new_lst.append(result[i])
    #print(new_lst)

    new_lst.sort()
    #print(new_lst)

    z = 0;
    for x in range(len(new_lst) - 1):
        if(new_lst[x] % new_lst[0] != 0):
            z = 1

    if(z == 0):
        print(new_lst[0])
    else:
        print(1)
    
        
```

```python
candies(['a', 'b', 'c', 'a', 'b', 'c', 'c', 'c'])
```
